## 4를 여러번 곱하기

4를 5번까지 곱해보자.

| 횟수 | 수식 | 결과 |
| --- |--- |:---|
|1|4|4|
|2|4x4|16|
|3|4x4x4|64|
|4|4x4x4x4|256|
|5|4x4x4x4x4|1024|

그리고 그려보자.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1,6)
y = 4**n

plt.plot(n,y,marker='o',lw=2,ls=':')
plt.xlabel('n')
```

## 여러번 곱할 걸 나눠서 곱하기

4를 2번 곱하기 = $4 \times 4 $

4를 3번 곱하기 = $4 \times 4 \times 4 = (4 \times 4) \times 4 = 4^2 \times 4$

일반화 하면 $ 4^n \times 4^m = 4^{(n+m)}$

## 정수가 아닌 횟수 곱하기
실제로 정수를 0.5번 곱할 수는 없음. 위에서 도출해낸 수식을 만족시키도록 개념을 확장만 해보자.

$ 4^n \times 4^m = 4^{(n+m)}$

$ n^{0.5} \times n^{0.5} = n^1 = n $

$n^{0.5}$ 는 자신을 두번 곱해서 n이 되는 수로 정의.

예를 들면 $4^{0.5} = 2, -2$

0.5제곱은 많이 쓰이므로 특별히 기호가 있다.

$\sqrt4 = 2$

1.2제곱 계산법: 1.2 = 12 / 10. 10번 곱하면 내가 되는 수를 찾아 (1/10) 그 수를 12번 곱|

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,5.1,0.1)
y = 4**n

plt.plot(n,y,marker='',lw=2,ls='-')
plt.xlabel('n')
```

## 미분해보기

| 횟수 | 수식 | 결과 |기울기|
| --- |--- |---|---|
|1|4|4|0
|2|4x4|16|12
|3|4x4x4|64|48
|4|4x4x4x4|256|192
|5|4x4x4x4x4|1024|768

그리고 그려보자.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n = np.array([2,3,4,5])
y = np.array([12,48,192,768])

plt.plot(n,y,marker='',lw=2,ls='-')
plt.xlabel('n')
```

```python
for y1,y2 in zip(y[0:],y[1:]):
    print('{} / {} = {}'.format(y2,y1,y2/y1))
```

기울기도 그대로 4배씩 증가. 여전히 4를 여러번 곱한게 된다. 미분해도 4의 지수는 4의 지수임. 미분하고 난담에 숫자는 스케일이 달라짐. 그렇다면 미분해서도 스케일이 달라지지 않는 경우가 있을까?

$$ \frac{d}{dx}e^x = e^x $$

$$ \frac{e^{(x+h)}-e^x }{h} = e^x$$

$$ e^{(x+h)}-e^x  = he^x$$

$$ e^x\cdot e^h  = (h+1) \cdot e^x$$

$$ e^h  = (h+1) $$

$$ e  = (h+1)^{1/h} $$
