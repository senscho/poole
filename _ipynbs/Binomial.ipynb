# binomial

probability of getting exactly n successes out of N ?
* success: p
* fail: q = 1-p

$ \frac{N!}{(N-n)!n!} \cdot p^n \cdot (1-p)^{N-n} $

```python
import numpy as np
from math import factorial

def fact(n):
    try:
        n[0]
    except:
        n = [n]
    return np.array([factorial(x) for x in n])
    
N = 10
p = 0.5
q = 1-p

n = np.arange(0,N+1)

y =fact(N)/(fact(N-n)*fact(n)) * p**n * q**(N-n)

import matplotlib.pyplot as plt

plt.bar(n,y)
plt.plot(n,y,'-',color='C1')
```
